```python
def unseen_brand(likes, input_date, inp_unseen_brand, k=5):
    input_date = datetime.strptime(input_date, '%Y-%m-%d %H:%M:%S')
    six_months = timedelta(days=30*6)
    start_date = input_date - six_months
    end_date = input_date
    final_df['date'] = pd.to_datetime(final_df['date'])
    
    filtered_df = final_df[(final_df['date'] >= start_date) & (final_df['date'] <= end_date)]
    
    if filtered_df.empty:
        print(f"No matching rows found for {input_date}")
        return None
    
    companies=[]
    for row in companies_arr:
        if inp_unseen_brand in row:
            companies=row
    
    filtered2_df = filtered_df[filtered_df['inferred company'].isin(companies)]
    
    filtered2_df['likes_distance'] = abs(filtered2_df['likes'] - likes)
    top_k_df = filtered2_df.nsmallest(k, 'likes_distance')
    
    return top_k_df.drop(columns=['likes_distance'])
```

```python
def unseen_period(input_company_name,input_likes,k=5):
    matching_rows = final_df[final_df['inferred company'] == input_company_name]

    if matching_rows.empty:
        print(f"No matching rows found for {input_company_name}")
        return None

    df_new = matching_rows.copy()

    df_new['likes_diff'] = abs(final_df['likes'] - input_likes)

    df_sorted = df_new.sort_values(by='likes_diff')

    if(len(df_sorted)>=k):
        return df_sorted[:k]
    else:
        return df_sorted
```

Go through each element of the dataframe. If the company is seen, then run unseen_period to generate the extra columns. If company is unseen, run unseen_brands to generate the extra columns. Those columns together with these are passed into the model for inference, we'll compare the output of the model with the content column.

```python
df = pd.read_csv("60k_df_Task2.csv")
```

```python
df['KNN 1'] = ""
df['KNN 2'] = ""
df['KNN 3'] = ""
```

```python
for i in range(len(df)):
    brand = df['inferred company'][i]
    if(i%500==0):    
        print(i)
    if brand in seen_brands:
        likes = df['likes'][i]
        df_output = unseen_period(brand,likes,4)
        quadruplet = df_output['content'].reset_index(drop=True).tolist()
        content = df['content'][i]
        # print(quadruplet)
        triplet = []
        for q in quadruplet:
            if(q!=content):
                triplet.append(q)
        # print(triplet)
        try:
            df['KNN 1'][i] = triplet[0]
            df['KNN 2'][i] = triplet[1]
            df['KNN 3'][i] = triplet[2]
        except:
            print(i)
            print(pd.Series(triplet))
        # break
    elif brand in unseen_brands:
        likes = df['likes'][i]
        date = df['date'][i]
        df_output = unseen_brands(likes, date, brand, 3)
        triplet = df_output['content'].reset_index(drop=True)
        df['KNN 1'][i] = triplet[0]
        df['KNN 2'][i] = triplet[1]
        df['KNN 3'][i] = triplet[2]
```
