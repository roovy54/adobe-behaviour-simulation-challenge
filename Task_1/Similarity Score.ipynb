```python
# CLIP
from tqdm import tqdm
from PIL import Image
import requests
import pandas as pd
from transformers import CLIPProcessor, CLIPModel
import urllib.request

model = CLIPModel.from_pretrained("openai/clip-vit-large-patch14").to("cuda")
processor = CLIPProcessor.from_pretrained("openai/clip-vit-large-patch14",device="cuda")

```

```python
# Calculating image-hypothesis similarity

def image_hypothesis_similarity(image_url,hypothesis):
  urllib.request.urlretrieve(image_url, 'img.jpg')

  image = Image.open('/content/img.jpg')

  inputs = processor(text=hypothesis, images=image, return_tensors="pt", padding=True).to('cuda')

  outputs = model(**inputs)
  logits_per_image = outputs.logits_per_image
  probs = logits_per_image.softmax(dim=1)

  return probs
```

```python
train=pd.read_csv("/content/drive/MyDrive/Adobe/final_data/test_new.csv")
```

```python
train.reset_index(inplace=True, drop=True)
```

```python
num_cols = 101
for i in range(num_cols):
    col_name = f'hyp_{i}'
    train[col_name] = '' * len(train)
```

```python
train.head()
```

```python
batch_size = 16

for i in tqdm(range(0, 10000, batch_size)):
    batch = train[i:i + batch_size]

    for idx, row in batch.iterrows():  # Iterate over each row in the batch
        try:
            if row["image_url"] != "Image does not exist":
                sim_score_image = image_hypothesis_similarity(row["image_url"], hypothesis)
                # Assign values from sim_score_image list to corresponding 'hyp_' columns
                for i, value in enumerate(sim_score_image[0]):
                    train.at[idx, f'hyp_{i}'] = value.item()
            else:
                train.at[idx, 'image_hypothesis_similarity'] = "Image does not exist"

            if row["video_thumbnail_url"] != "Video thumbnail does not exist":
                sim_score_video_thumbnail = image_hypothesis_similarity(row["video_thumbnail_url"], hypothesis)
                # Assign values from sim_score_video_thumbnail list to corresponding 'hyp_' columns
                for i, value in enumerate(sim_score_video_thumbnail[0]):
                    train.at[idx, f'hyp_{i}'] = value.item()
            else:
                train.at[idx, 'video_thumbnail_hypothesis_similarity'] = "Video thumbnail does not exist"

        except Exception as e:
            print(f"Error processing row {idx}: {e}")
            # Handle exceptions as needed
```
