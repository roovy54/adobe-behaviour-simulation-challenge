```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train_df = pd.read_csv("train_full.csv")
```

```python
train_df
```

```python
tr_df_sorted = train_df.sort_values(by='likes')
```

```python
df_spotify = tr_df_sorted[tr_df_sorted['inferred company'] == 'spotify']
plt.hist(df_spotify.index, bins=20, edgecolor='black')

plt.xlabel('Index of DataFrame')
plt.ylabel('Number of Spotify Tweets')
plt.title('Distribution of Spotify Tweets Across the Dataset')

plt.show()
```
