```python
import cv2
import numpy as np
import torchvision
from torchvision import transforms
import pytorchvideo
```

```python
cv2_vr = cv2.VideoCapture("/content/eaqkEB1zjKN1u2xv.mp4")
duration = int(cv2_vr.get(cv2.CAP_PROP_FRAME_COUNT))
frame_id_list = np.linspace(0, duration-1, 8, dtype=int)

video_data = []
for frame_idx in frame_id_list:
    cv2_vr.set(1, frame_idx)
    _, frame = cv2_vr.read()
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    video_data.append(torch.from_numpy(frame).permute(2, 0, 1))
cv2_vr.release()
video_data = torch.stack(video_data, dim=1)
video_outputs = transform(video_data)
```

```python
import cv2

video_path = '/content/eaqkEB1zjKN1u2xv.mp4'  # Replace with your video file path
vidcap = cv2.VideoCapture(video_path)

success, image = vidcap.read()
count = 0

while success:
    cv2.imwrite(f"frame_{count}.png", image)  # Save each frame as an image
    success, image = vidcap.read()
    print(f"Frame {count} extracted")
    count += 1
```

```python
import os

# Directory containing the images
directory_path = '/content/selectedframes'  # Replace this with your directory path

# List to store image paths
image_paths = []

# Iterate through the directory
for root, dirs, files in os.walk(directory_path):
    for file in files:
        # Check if the file has an image extension (e.g., jpg, png, etc.)
        if file.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp')):
            # Construct the full path to the image file
            image_path = os.path.join(root, file)
            image_paths.append(image_path)

# Print the list of image paths
print(image_paths)
```
